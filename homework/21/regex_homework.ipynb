## Testing single patterns

```python
text = open("dates.txt", "r").read()
```

```python
pat_1 = r'\b(?:0[1-9]|1[0-2])/(?:0[1-9]|[12]\d|3[01])/(?:\d{2})?\d{2,4}\b'

dat_1 = re.findall(pat_1, text)

#'Janaury 1993' in dat_1
```

```python
pat_2 = r'\b(?:\d{1,2} )?(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*(?: \d{1,2}[ ,])? (?:\d{2})?\d{2,4}\b'

dat_2 = re.findall(pat_2, text)

#'Janaury 1993' in dat_2
```

```python
pat_3 = r'\b(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* \d{4}\b'

dat_3 = re.findall(pat_3, text)

#'24/1990' in dat_3
```

```python
pat_4 = r'\b(?:0[1-9]|1[0-2])/\d{4}\b'

dat_4 = re.findall(pat_4, text)

#'24/1990' in dat_4
```

```python
pat_5 = r'\b\d{4}\b'

dat_5 = re.findall(pat_5, text)

#'9/36/308' in dat_5
```

```python
dates = dat_1 + dat_2 + dat_3 + dat_4 + dat_5
```

```python
import pandas as pd

def parse_dates_in_text_to_df(text):
    lines = text.split('\n')
    #combined pattern
    date_pattern = r'\b(?:\d{1,2}/\d{1,2}/(?:\d{2})?\d{2,4}|(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[- ,.]? \d{1,2}(?:st|nd|rd|th)?,? \d{2,4}|(?:0[1-9]|1[0-2])/\d{4}|(?:\d{1,2} )?(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*(?: \d{1,2}[ ,])? (?:\d{2})?\d{2,4})\b'
    
    parsed_dates = []
    for idx, line in enumerate(lines, start=0):
        dates_in_line = re.findall(date_pattern, line)
        for date_str in dates_in_line:
            try:
                date = parser.parse(date_str, fuzzy=True)
                parsed_dates.append((date, idx))
            except ValueError:
                pass  # Ignore errors and continue
    
    df = pd.DataFrame(parsed_dates, columns=['date', 'index'])
    return df
```

```python
df = parse_dates_in_text_to_df(text)
len(df)
```
