```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
dataset= df= pd.read_csv('Groceries_dataset.csv',)
df.head()
```

```python
df.shape
```

```python
df['Member_number'] = df['Member_number'].astype(str)
```

```python
df.loc[(df['Member_number'] == '1808') & (df["Date"] == '29-11-2014')]
```

```python
df['itemDescription'].value_counts()
```

```python
#Group data
group_items = df.groupby(['Member_number', 'Date'])['itemDescription'].unique().reset_index()
group_items
```

```python
#data must be lists in list type
new_data = []

for index, row in group_items.iterrows():
    items = row['itemDescription'].tolist()
    new_data.append(items)
    
new_data    
```

```python
rules = rules = list(apriori(new_data, min_support=0.001, min_confidence=0.1, min_lift=1, min_length=1))

rules
```

```python
len(rules)
```

```python
def print_rules(rules):
    for rule in rules:
        print ('rule.items=', list(rule.items))
        print ('rule.support=',rule.support)

        for os in rule.ordered_statistics:
            print ('\titems_base=', list(os.items_base))
            print ('\tlifted_item =', list(os.items_add))
            print ('\tlift=', os.lift)
            print ('\tconfidence (i.e. cond prob {} if {})='.format(list(os.items_add), list(os.items_base)), os.confidence)
            print ('\n')
        
print_rules(rules)
```

## Answer

```python
top_10_rules = sorted(rules, key=lambda x: (x.support, x.ordered_statistics[0].confidence, x.ordered_statistics[0].lift), reverse=True)[:10]


for rule in top_10_rules:
    print(f"Items: {rule.items}, Support: {rule.support:.4f}, Confidence: {rule.ordered_statistics[0].confidence:.4f}, Lift: {rule.ordered_statistics[0].lift:.4f}")
```
