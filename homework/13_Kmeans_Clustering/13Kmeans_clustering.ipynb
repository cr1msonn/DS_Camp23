```python
from sklearn.datasets import make_blobs
import numpy as np 
import matplotlib.pyplot as plt

```

```python
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 5, 
                           cluster_std = 0.6, random_state = 0)
colors = np.array([plt.cm.Spectral(val)
          for val in np.linspace(0, 1, len(set(y)))])
plt.figure(figsize=(8,6))

plt.scatter(X[:,0], X[:,1], c= colors[y], s= 20)  
```

## Agglomerative clustering using single linkage

```python
from sklearn.cluster import AgglomerativeClustering
clf = AgglomerativeClustering(n_clusters=5, linkage= 'single')

predicted = clf.fit_predict(X)

plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:, 1], c= colors[predicted], marker= '.') 
plt.title ('sklearn kmeans single linkage')
```

## Agglomerative clustering using complete linkage

```python
from sklearn.cluster import AgglomerativeClustering
clf = AgglomerativeClustering(n_clusters=5, linkage= 'complete')

predicted = clf.fit_predict(X)

plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:, 1], c= colors[predicted], marker= '.') 
plt.title ('sklearn kmeans complete linkage')
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps = .5, min_samples = 5)

predicted = dbscan.fit_predict(X)

plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:, 1], c= colors[predicted], marker= '.') 
plt.title ('sklearn kmeans complete linkage')
```

## DBSCAN with diff parameters

```python
dbscan = DBSCAN(eps = .8, min_samples = 3)

predicted = dbscan.fit_predict(X)

plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:, 1], c= colors[predicted], marker= '.') 
plt.title ('sklearn kmeans dbscan')
```

## Elbow Method for KMeans

```python
from sklearn.cluster import KMeans

scores = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit_predict(X)
    scores.append(kmeans.inertia_)

plt.plot(range(1, 10), scores, marker='o')
plt.title('Elbow Method for k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Score')
plt.show()
```

#### We can clearly see that after 5th cluster the score is slowing down, so it will be the perfect k for our data

```python
#Using k that was chosen by elbow method
predicted = KMeans(n_clusters=5).fit_predict(X)

plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:, 1], c= colors[predicted], marker= '.') 
plt.title ('sklearn kmeans dbscan')
```
