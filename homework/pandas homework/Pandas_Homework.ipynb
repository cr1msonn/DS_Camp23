## Answer 1

```python
def answer_one():

    import numpy as np
    import pandas as pd

    energy = pd.read_excel('Energy Indicators.xls', skiprows= 17, skipfooter=38)
    # Drop unnecessary columns
    energy.drop(energy.columns[[0,1]], axis = 1 , inplace = True)
    # Rename the columns
    energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    # Replacing missing values (set to ...) with np.Nan
    energy.replace('...', np.NaN, inplace=True)
    # Converting Energy Supply to gigajoules (there are 1,000,000 gigajoules in a petajoule).
    energy['Energy Supply'] = energy['Energy Supply'].mul(10**6)

    
    def digit_parant_clean(data):

        if data.find('(') > -1:
            data = data[:data.find('(')]   
        data = ''.join([x for x in data if not x.isdigit()])
        
        return data.strip()
    
    # Cleaning the data by removing parentheses and numbers
    energy['Country'] = energy['Country'].apply(digit_parant_clean)
    
    country_rename = {"Republic of Korea" : "South Korea", 
                  "United States of America" : "United States",
                  "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
                  "China, Hong Kong Special Administrative Region": "Hong Kong"}
    # Renaming countries
    energy.replace({"Country": country_rename},inplace = True)

    gdp = pd.read_csv('world_bank.csv', skiprows=4)
    gdp.rename(columns =  {"Country Name":"Country"}, inplace = True)
    cr  = {"Korea, Rep." : "South Korea",
        "Iran, Islamic Rep.":"Iran",
        "Hong Kong SAR, China": "Hong Kong"}
    # Renaming countries
    gdp.replace({"Country": cr}, inplace = True)
    columnstokeep = ['Country', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    gdp = gdp[columnstokeep]
    # Reading excel file and keeping only 15 rows (rank 1-15), also dropping unnecessary column
    ScimEn = pd.read_excel("country_rank.xlsx", nrows=15).drop(["Region"], axis=1)
    # Merging all df`s together
    df = pd.merge(pd.merge(ScimEn, energy, on='Country'), gdp, on='Country')
    #The index of this DataFrame should be the name of the country
    df.set_index('Country',inplace=True)

    return df

one = answer_one()
one
      
```

## Answer 2

```python
def answer_two():
    #Get the df from question 1
    Top15 = answer_one()
    
    # Calculating the average GDP over the last 10 years for each country and sort it in descending order
    avgGDP = Top15[Top15.columns[10:]].mean(axis=1).sort_values(ascending=False)

    #return Series of average GDP values
    return avgGDP

print(answer_two())
```

## Answer 3

```python
def answer_three():
    #Get the df from question 1
    Top15 = answer_one()
    # Get a name of the country with the 6th largest average GDP (.index[5]) (since i cant use answer_two() function i do the same as i did there)
    six_largest_gdp_country = Top15[Top15.columns[10:]].mean(axis=1).sort_values(ascending=False).index[5]

    # Calculating how much had the GDP changed over the 10 year span for the country with the 6th largest average GDP
    span = Top15.loc[six_largest_gdp_country]['2015'] - Top15.loc[six_largest_gdp_country]['2006']

    return span

print(answer_three())
```

```python
def answer_four():
    #Get the df from question 1
    Top15  = answer_one()
    # Calculate the ratio
    Top15["Self-citations to Citations ratio"] = Top15['Self-citations'] / Top15['Citations']
    
    # Return the tuple with the name of the country with max ratio and it`s ratio
    return (Top15["Self-citations to Citations ratio"].idxmax(), Top15["Self-citations to Citations ratio"].max())

print(answer_four())
```

## Answer 5

```python
def answer_five():
    #Get the df from question 1
    Top15 = answer_one()

    #Calculate the population based on the energy supply per capita
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']

    #Return the country name with max population
    return Top15['Population'].idxmax()

print(answer_five())
```

```python

def answer_six():
    #Get the df from question 1
    Top15 = answer_one()
    #Calculate the population based on the energy supply per capita again 
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    #Calculate the Citable Documents per person based on the population
    Top15['Citable Documents per person'] = Top15['Citable documents'] / Top15['Population']
    
    return Top15['Citable Documents per person'].corr(Top15['Energy Supply per Capita'])
    
print(answer_six())
```

## Answer 7

```python
def answer_seven():
    #Get the df from question 1
    Top15 = answer_one()

    #Calculate the population based on the energy supply per capita again 
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America',
                  }
    
    #Turning dict to series object so indices with Top15 will match
    ContinentDict= pd.Series(ContinentDict)

    #Assigning new column "Continent" based on country
    Top15 = Top15.assign(Continent = ContinentDict)

    #Grouping by Continent and using only Population column
    grouped_population = Top15.groupby('Continent')['Population']

    #New DataFrame with all metrics we are interested in
    newdf = grouped_population.agg(['count', 'sum', 'mean', 'std'])

    #Renaming the columns
    newdf.columns = ['size', 'sum', 'mean', 'std']
    #Dealing with missing values
    newdf['std'].fillna(0, inplace= True)
    
    return newdf

seven = answer_seven()
seven
```
