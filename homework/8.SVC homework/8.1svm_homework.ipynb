### Sklearn SVM for Iris dataset


```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
print ('data contains:',iris.keys())
X, y, labels, feature_names  = iris.data, iris.target, iris.target_names, iris['feature_names']
df_iris= pd.DataFrame(X, columns= feature_names) 
df_iris['label'] =  y
features_dict = {k:v for k,v in  enumerate(labels)}
df_iris['label_names'] = df_iris.label.apply(lambda x: features_dict[x])
df_iris
```

### Sklearn LinearSVC

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(X,y, test_size=0.2, random_state = 0)

X_train.shape, X_test.shape
```

```python
from sklearn.svm import LinearSVC
```

```python
clf = LinearSVC(C=1).fit(X_train, y_train)
print("train accuracy= {:.3%}".format(clf.score(X_train, y_train)))
print  ('b = {}\nw = {}'.format(clf.intercept_, clf.coef_))
```

```python
from sklearn.svm import SVC

iris = load_iris()
X = iris.data[:, :2]
y = iris.target

svc = LinearSVC(C=1)
svc.fit(X, y)


h = .02  
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)


plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', s=20)
plt.title('SVC Linear')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
from sklearn.svm import SVC
degree= 3
clf = SVC(kernel= 'poly').fit(X_train,y_train) # default=3)
print("train accuracy= {:.3%}".format(clf.score (X_train,y_train)))

iris = load_iris()
X = iris.data[:, :2]
y = iris.target

svc = SVC(kernel= 'poly')
svc.fit(X, y)


h = .02  
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)


plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', s=20)
plt.title('SVC Poly')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
gamma=1
clf = SVC(C=1, gamma=gamma).fit(X_train,y_train) 
print("train accuracy= {:.3%}".format(clf.score (X_train,y_train)))

iris = load_iris()
X = iris.data[:, :2]
y = iris.target

svc = SVC(C=1, kernel = 'rbf', gamma = gamma)
svc.fit(X, y)


h = .02  
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)


plt.contourf(xx, yy, Z, alpha=0.8,)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', s=20)
plt.title('SVC Rbf with gamma')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```
